# Notebook to plot training progress of agent

It will obtain data from file stored in `bullet_racecar.cma.1.32.hist.json`, where format is `env_name.optimizer.num_rollouts.popsize.hist.json`.


```python
import numpy as np
import json
import os
import matplotlib.pyplot as plt
```

```python
#Compare population 16 to 32
env_name = 'bipedeasyphys'
optimizer0 = 'oes'
optimizer1 = 'oes'
optimizer2 = 'experimentGrafting'
optimizer3 = 'experimentGrafting'
num_rollouts = 16 # number of rollouts that are averaged over an episode
popsize = 16

file_base0 = env_name+'.'+optimizer0+'.'+str(num_rollouts)+'.'+str(popsize)
filename0 = os.path.join('log', file_base1+'.hist.json')

file_base1 = env_name+'.'+optimizer1+'.'+str(num_rollouts)+'.'+str(32)
filename1 = os.path.join('log', file_base1+'.hist.json')

file_base2 = env_name+'.'+optimizer2+'.'+str(num_rollouts)+'.'+str(popsize)
filename2 = os.path.join('log', file_base2+'.hist.json')

file_base3 = env_name+'.'+optimizer3+'.'+str(num_rollouts)+'.'+str(popsize)
filename3 = os.path.join('log', file_base3+'.hist.json')

with open(filename0, 'r') as f:
  raw_data = json.load(f)
data0 = np.array(raw_data)

with open(filename1, 'r') as f:
  raw_data = json.load(f)
data1 = np.array(raw_data)

with open(filename2, 'r') as f:
  raw_data = json.load(f)
data2 = np.array(raw_data)

with open(filename3, 'r') as f:
  raw_data = json.load(f)
data3 = np.array(raw_data)

fig=plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
data0_max, = plt.plot(data0[:, 0]/60., data0[:, 2])
data1_max, = plt.plot(data1[:, 0]/60., data1[:, 2])
#data2_max, = plt.plot(data2[:, 0]/60., data2[:, 4])
#data3_max, = plt.plot(data3[:, 0]/60., data3[:, 4])
#line_min, = plt.plot(data1[:, 1]/60., data1[:, 3])
#line_max, = plt.plot(data1[:, 1]/60., data1[:, 4])
plt.legend([data0_max,data1_max], ['Hybrid Max w/ Population of 16', 'Hybrid Max w/ Population of 32'])
plt.xlabel('generation')
plt.ylabel('cumulative reward')
plt.title('Comparing Results based on Population Size')
#plt.savefig(file_base1+".svg")
plt.show()
```
